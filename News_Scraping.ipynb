```python
import requests
from bs4 import BeautifulSoup
from rich.console import Console
from rich.table import Table
from urllib.parse import urlparse

HEADERS = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}
console = Console()

def scrape_news(url):
    try:
        response = requests.get(url, headers=HEADERS, timeout=10)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        console.print(f"[bold red]âŒ Error fetching data from {url}: {e}[/bold red]")
        return

    soup = BeautifulSoup(response.text, 'html.parser')
    headlines = set()

    for article in soup.find_all("article"):
        headline_tag = article.find(["h2", "h3", "h4"])
        if headline_tag:
            text = headline_tag.get_text(strip=True)
            link_tag = article.find("a", href=True)
            link = f"https://{urlparse(url).netloc}{link_tag['href']}" if link_tag else "No URL"
            if text and "Deals" not in text and "Games" not in text:
                headlines.add((headline_tag.name, text, link))

    if not headlines:
        console.print(f"[bold bright_yellow]âš ï¸ No relevant headlines found on {url}.[/bold bright_yellow]")
        return

    site_name = urlparse(url).netloc.replace("www.", "").replace(".com", "").upper()
    table = Table(title=f"ğŸ”¥ {url} HEADLINES", header_style="bold bright_magenta", show_lines=True, show_header=True,safe_box=True)
    table.add_column("Website", justify="left", style="bright_green",min_width=15, max_width=25)
    table.add_column("Headline", justify="left", style="bright_yellow")
    table.add_column("Link", justify="left", style="bright_cyan", overflow="ellipsis",max_width=100)

    for tag, text, link in list(headlines)[:10]:
        table.add_row(site_name, text, link)

    console.print(table)

news_urls = [
    "https://www.foxnews.com/entertainment",
    "https://www.foxnews.com/media",
    "https://www.foxnews.com/sports",
    "https://www.foxnews.com/world",
    "https://www.foxnews.com/tech",
    "https://www.foxnews.com/business",
    "https://www.foxnews.com/politics",
    "https://www.foxnews.com/health"
]

for url in news_urls:
    scrape_news(url)
```
